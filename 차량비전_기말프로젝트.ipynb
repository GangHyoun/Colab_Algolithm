```python
# 두 이미지 읽기
img1 = imgk1
img2 = imgk2

# SIFT 검출기 생성
sift = cv2.SIFT_create()

# 키포인트와 디스크립터 검출
keypoints1, descriptors1 = sift.detectAndCompute(img1, None)
keypoints2, descriptors2 = sift.detectAndCompute(img2, None)

# BFMatcher로 매칭
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
matches = bf.match(descriptors1, descriptors2)

# 매칭된 키포인트 정렬
matches = sorted(matches, key=lambda x: x.distance)

# 상위 n개 매칭점 선택
n_matches = 10
matches = matches[:n_matches]

# 매칭된 좌표 추출
pts1 = np.float32([keypoints1[m.queryIdx].pt for m in matches]).reshape(-1, 1, 2)
pts2 = np.float32([keypoints2[m.trainIdx].pt for m in matches]).reshape(-1, 1, 2)

# Fundamental Matrix 계산
F, mask = cv2.findFundamentalMat(pts1, pts2, cv2.FM_RANSAC)

# 선택된 inlier 포인트
pts1_inliers = pts1[mask.ravel() == 1]
pts2_inliers = pts2[mask.ravel() == 1]

print("Fundamental Matrix:")
print(F)

# 매칭된 포인트 그리기
img_matches = cv2.drawMatches(img1, keypoints1, img2, keypoints2, matches, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# 이미지 표시
plt.figure(figsize=(20, 10))
plt.imshow(img_matches)
plt.axis("off")
plt.show()
```

```python
# 이미지 로드 (흑백으로 로드)
left_img = imge1
right_img = imge2

# Block Matching 파라미터 설정
block_size = 15  # 블록 크기
min_disparity = 0  # 최소 disparity
max_disparity = 16  # 최대 disparity
num_disparities = max_disparity - min_disparity  # disparity 범위

# Block Matching 알고리즘 생성
stereo = cv2.StereoBM_create(numDisparities=num_disparities, blockSize=block_size)

# 좌우 이미지에 대해 disparity 계산
disparity = stereo.compute(left_img, right_img)

# disparity를 실수형으로 변환하여 정규화
disparity = disparity.astype(np.float32) / 16.0

# disparity를 시각화
plt.imshow(disparity, cmap='gray')
#plt.title('Disparity Map')
plt.axis("off")
plt.show()
```

```python
img_left = imge1
img_right = imge2

# SGBM 객체 생성 및 매개변수 설정
window_size = 5  # 블록 크기. 홀수여야 함.
min_disp = 0  # 최소 디스패리티
num_disp = 16 * 5  # 최대 디스패리티. 16의 배수

stereo = cv2.StereoSGBM_create(minDisparity=min_disp,
                               numDisparities=num_disp,
                               blockSize=window_size,
                               P1=8 * 3 * window_size ** 2,
                               P2=32 * 3 * window_size ** 2,
                               disp12MaxDiff=1,
                               uniquenessRatio=15,
                               speckleWindowSize=100,
                               speckleRange=32,
                               preFilterCap=63,
                               mode=cv2.STEREO_SGBM_MODE_SGBM_3WAY)

# 깊이 맵 계산
disparity = stereo.compute(img_left, img_right).astype(np.float32) / 16.0

# 깊이 맵을 시각적으로 보기 위해 정규화
disparity = cv2.normalize(disparity, disparity, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
disparity = np.uint8(disparity)

# 결과 출력
plt.imshow(disparity, cmap='gray')
#plt.title('Disparity Map')
plt.axis("off")
plt.show()
```

```python
import torch
from torchvision.transforms import functional as F
from PIL import Image

# MiDaS 모델 불러오기
model = torch.hub.load("intel-isl/MiDaS", "MiDaS")

# 이미지 불러오기
img_left = cv2.imread('/content/drive/MyDrive/image_final/middelburry/ime0.png')
img_right = cv2.imread('/content/drive/MyDrive/image_final/middelburry/ime1.png')

# 이미지를 RGB로 변환
img_left_rgb = cv2.cvtColor(img_left, cv2.COLOR_BGR2RGB)
img_right_rgb = cv2.cvtColor(img_right, cv2.COLOR_BGR2RGB)

# 이미지를 PIL 포맷으로 변환
img_left_pil = Image.fromarray(img_left_rgb)
img_right_pil = Image.fromarray(img_right_rgb)


# 이미지 크기를 MiDaS 모델이 요구하는 크기인 384x384로 조정
img_left_resized = F.resize(img_left_pil, (384, 384))
img_right_resized = F.resize(img_right_pil, (384, 384))

# 이미지를 모델 입력 형식에 맞게 변환
img_left_input = F.to_tensor(img_left_resized).unsqueeze(0).cuda()
img_right_input = F.to_tensor(img_right_resized).unsqueeze(0).cuda()

# 모델을 evaluation 모드로 설정
model.eval()

# 모델을 GPU로 이동
model.cuda()

# 이미지를 GPU로 옮기고 예측 수행
with torch.no_grad():
    depth_map = model(img_left_input)

# Depth map을 시각화하기 위해 numpy 배열로 변환
depth_map_np = depth_map.squeeze().cpu().numpy()

# 결과 시각화
plt.imshow(depth_map_np, cmap='rainbow')  # 다른 색상 맵을 선택하려면 cmap 옵션을 변경할 수 있습니다.
#plt.title('Depth Map (MiDaS)')
plt.axis("off")
plt.show()
```
