1번

*   위 세개의 필터의 사이즈를 3 by 3으로 설정 후 노이즈 제거 결과
*   위 세개의 필터의 사이즈를 7 by 7으로 설정 후 노이즈 제거 결과
*   위 세개의 필터의 사이즈를 11 by 11으로 설정 후 노이즈 제거 결과
*   위 세개의 필터의 사이즈를 17 by 17으로 설정 후 노이즈 제거 결과

*단, Gaussian filter 의 분산값은 1.0으로 설정

```python
# 이미지 불러오기
imgF1 = cv2.imread('/content/drive/MyDrive/image_openCV/Lena-Gaussian-noise.jpg')
imgF2 = cv2.imread('/content/drive/MyDrive/image_openCV/fce(salt_pepper noise).bmp')

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(imgF1)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Original")
plt.imshow(imgF2)
plt.axis("off")
```

```python
# Average Filter 생성 및 kernel 사이즈 설정

# 3, 7, 11, 17 평균 필터 커널 생성
# nxn 필터의 요소 개수(n^2)로 나누어줌
k3 = np.ones((3,3))/3**2
k7 = np.ones((7,7))/7**2
k11 = np.ones((11,11))/11**2
k17 = np.ones((17,17))/17**2
```

```python
# Average Filter image1

# 필터 적용
b3 = cv2.filter2D(imgF1, -1, k3)
b7 = cv2.filter2D(imgF1, -1, k7)
b11 = cv2.filter2D(imgF1, -1, k11)
b17 = cv2.filter2D(imgF1, -1, k17)

plt.subplot(1,5,1)
plt.title("Original")
plt.imshow(imgF1)
plt.axis("off")

plt.subplot(1,5,2)
plt.title("Avrg 3x3")
plt.imshow(b3)
plt.axis("off")

plt.subplot(1,5,3)
plt.title("Avrg 7x7")
plt.imshow(b7)
plt.axis("off")

plt.subplot(1,5,4)
plt.title("Avrg 11x11")
plt.imshow(b11)
plt.axis("off")

plt.subplot(1,5,5)
plt.title("Avrg 17x17")
plt.imshow(b17)
plt.axis("off")
```

```python
# Average Filter image2
# 사용법은 필터코드 참고자료 참고

# 필터 적용
b3 = cv2.filter2D(imgF2, -1, k3)
b7 = cv2.filter2D(imgF2, -1, k7)
b11 = cv2.filter2D(imgF2, -1, k11)
b17 = cv2.filter2D(imgF2, -1, k17)

plt.subplot(1,5,1)
plt.title("Original")
plt.imshow(imgF2)
plt.axis("off")

plt.subplot(1,5,2)
plt.title("Avrg 3x3")
plt.imshow(b3)
plt.axis("off")

plt.subplot(1,5,3)
plt.title("Avrg 7x7")
plt.imshow(b7)
plt.axis("off")

plt.subplot(1,5,4)
plt.title("Avrg 11x11")
plt.imshow(b11)
plt.axis("off")

plt.subplot(1,5,5)
plt.title("Avrg 17x17")
plt.imshow(b17)
plt.axis("off")
```

```python
# Median Filter image1
# 사용법은 필터코드 참고자료 참고
# dst = cv2.medianBlur(src, ksize)
# src: 입력 영상, ksize: 커널 크기

# 미디언 블러 적용
b3 = cv2.medianBlur(imgF1, 3)
b7 = cv2.medianBlur(imgF1, 7)
b11 = cv2.medianBlur(imgF1, 11)
b17 = cv2.medianBlur(imgF1, 17)

plt.subplot(1,5,1)
plt.title("Original")
plt.imshow(imgF1)
plt.axis("off")

plt.subplot(1,5,2)
plt.title("Medi 3x3")
plt.imshow(b3)
plt.axis("off")

plt.subplot(1,5,3)
plt.title("Medi 7x7")
plt.imshow(b7)
plt.axis("off")

plt.subplot(1,5,4)
plt.title("Medi 11x11")
plt.imshow(b11)
plt.axis("off")

plt.subplot(1,5,5)
plt.title("Medi 17x17")
plt.imshow(b17)
plt.axis("off")
```

```python
# Median Filter image2
# dst = cv2.medianBlur(src, ksize)
# src: 입력 영상, ksize: 커널 크기

# 미디언 블러 적용
b3 = cv2.medianBlur(imgF2, 3)
b7 = cv2.medianBlur(imgF2, 7)
b11 = cv2.medianBlur(imgF2, 11)
b17 = cv2.medianBlur(imgF2, 17)

plt.subplot(1,5,1)
plt.title("Original")
plt.imshow(imgF2)
plt.axis("off")

plt.subplot(1,5,2)
plt.title("Medi 3x3")
plt.imshow(b3)
plt.axis("off")

plt.subplot(1,5,3)
plt.title("Medi 7x7")
plt.imshow(b7)
plt.axis("off")

plt.subplot(1,5,4)
plt.title("Medi 11x11")
plt.imshow(b11)
plt.axis("off")

plt.subplot(1,5,5)
plt.title("Medi 17x17")
plt.imshow(b17)
plt.axis("off")
```

위 6개의 필터를 본인이 선택한 이미지에 모두 적용한 결과 이미지


```python
# 이미지 두 개 선택 후 그레이 스케일로 변경
# 1번 이미지는 잡음이 섞인 이미지
imgE1 = cv2.imread('/content/drive/MyDrive/image_openCV/Lena-Gaussian-noise.jpg', cv2.IMREAD_GRAYSCALE)
#imgE1 = cv2.imread('/content/drive/MyDrive/image_openCV/나몰빼미_프사_배잠.png', cv2.IMREAD_GRAYSCALE)

imgE2 = cv2.imread('/content/drive/MyDrive/image_openCV/cat.bmp', cv2.IMREAD_GRAYSCALE)
#imgE2 = cv2.imread('/content/drive/MyDrive/image_openCV/글레SD.png', cv2.IMREAD_GRAYSCALE)

plt.subplot(1,2,1)
plt.title('Gray 1')
plt.imshow(imgE1, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title('Gray 2')
plt.imshow(imgE2, cmap="gray")
plt.axis("off")
```

```python
# 로버트 크로스
# 로버트 커널 생성
gx_kernel = np.array([[1,0], [0,-1]])
gy_kernel = np.array([[0, 1],[-1,0]])

# 커널 적용
edge_gx1 = cv2.filter2D(imgE1, -1, gx_kernel)
edge_gy1 = cv2.filter2D(imgE1, -1, gy_kernel)

edge_gx2 = cv2.filter2D(imgE2, -1, gx_kernel)
edge_gy2 = cv2.filter2D(imgE2, -1, gy_kernel)

outputImgRo1 = edge_gx1+edge_gy1
outputImgRo2 = edge_gx2+edge_gy2

plt.subplot(1,2,1)
plt.title('Robert 1')
plt.imshow(outputImgRo1, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title('Robert 2')
plt.imshow(outputImgRo2, cmap="gray")
plt.axis("off")
```

```python
#소벨
#dx와 dy를 합쳐 출력
#delta값을 정해주지 않으면 미분이 - 부분은 0
# x -> depth = -1, dx = 1, dy = 0
# y -> depth = -1, dx = 0, dy = 1
dx1 = cv2.Sobel(imgE1, -1, 1, 0, delta = 128)
dy1 = cv2.Sobel(imgE1, -1, 0, 1, delta = 128)

dx2 = cv2.Sobel(imgE2, -1, 1, 0, delta = 128)
dy2 = cv2.Sobel(imgE2, -1, 0, 1, delta = 128)

#uint8 타입(0~255)로 변경하고 영상 합치기
absX1 = cv2.convertScaleAbs(dx1)
absY1 = cv2.convertScaleAbs(dy1)

absX2 = cv2.convertScaleAbs(dx2)
absY2 = cv2.convertScaleAbs(dy2)

outputImg1 = cv2.addWeighted(absX1, 0.5, absY1, 0.5, 0)

outputImg2 = cv2.addWeighted(absX2, 0.5, absY2, 0.5, 0)

plt.subplot(1,2,1)
plt.title('Sobel 1')
plt.imshow(outputImg1, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title('Sobel 2')
plt.imshow(outputImg2, cmap="gray")
plt.axis("off")
```

```python
#프리윗
# 프리윗 커널 생성
gx_k = np.array([[-1,0,1], [-1,0,1],[-1,0,1]])
gy_k = np.array([[-1,-1,-1],[0,0,0], [1,1,1]])

# 프리윗 커널 필터 적용
edge_gx1 = cv2.filter2D(imgE1, -1, gx_k)
edge_gy1 = cv2.filter2D(imgE1, -1, gy_k)

edge_gx2 = cv2.filter2D(imgE2, -1, gx_k)
edge_gy2 = cv2.filter2D(imgE2, -1, gy_k)

outputImgPre1 = edge_gx1+edge_gy1
outputImgPre2 = edge_gx2+edge_gy2

plt.subplot(1,2,1)
plt.title('Prewitt 1')
plt.imshow(outputImgPre1, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title('Prewitt 2')
plt.imshow(outputImgPre2, cmap="gray")
plt.axis("off")
```

```python
#캐니
#캐니 에지 컨벌루션 연산, T1=200, T2=100
outputImgC1 = cv2.Canny(imgE1, 100,200)
outputImgC2 = cv2.Canny(imgE2, 100,200)

plt.subplot(1,2,1)
plt.title('Canny 1')
plt.imshow(outputImgC1, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title('Canny 2')
plt.imshow(outputImgC2, cmap="gray")
plt.axis("off")
```

```python
#라플라시안
#라플라시안 에지 컨벌루션 연산
# depth = -1, ksize = 1 (3x3)
laplacian1 = cv2.Laplacian(imgE1,-1,1)

laplacian2 = cv2.Laplacian(imgE2,-1,1)

plt.subplot(1,2,1)
plt.title('Laplacian 1')
plt.imshow(laplacian1, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title('Laplacian 2')
plt.imshow(laplacian2, cmap="gray")
plt.axis("off")
```

```python
#LoG (Laplacian of Gaussian)
#Gaussian
blur1 = cv2.GaussianBlur(imgE1,(3,3),1)
#Laplacian
outputImg1 = cv2.Laplacian(blur1, -1,1)

#Gaussian
blur2 = cv2.GaussianBlur(imgE2,(3,3),1)
#Laplacian
outputImg2 = cv2.Laplacian(blur2, -1,1)

plt.subplot(1,2,1)
plt.title('LoG 1')
plt.imshow(outputImg1, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.title('LoG 2')
plt.imshow(outputImg2, cmap="gray")
plt.axis("off")
```
